```python
data=pd.read_csv(r"C:\Users\HP\Downloads\vehicles.csv")
```

```python
data
```

```python
data.columns
```

```python
data = data.rename(columns={'Make':'Producer',
                            'Engine Displacement':'Displacement'})
```

```python
data.columns
```

```python
column_order = ['Year','Vehicle Class','Transmission','Drivetrain','Fuel Type',
                'Cylinders','Displacement','Fuel Barrels/Year',
                'City MPG','Highway MPG','Combined MPG',
                'CO2 Emission Grams/Mile','Fuel Cost/Year']


data = data[column_order]

```

```python
data
```

```python
filtered = data[(data['Fuel Type']=='Regular') & 
                (data['Cylinders']>=6) & 
                (data['Fuel Cost/Year'] < 2000)]


filtered.head()
```

```python
avg_mpg = data.groupby('Fuel Type', as_index=False).agg({'Combined MPG':'mean'})
avg_mpg.columns = ['Fuel Type', 'Avg_MPG']


```

```python
avg_mpg
```

```python

data = pd.merge(data, avg_mpg, on='Fuel Type')
data.head(10)
```
