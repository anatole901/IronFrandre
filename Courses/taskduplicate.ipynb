```python
accounts = list(data['Account'].unique())
accounts
```

```python
for orga in accounts :
    if data.loc[data['Account'] == orga].shape[0] > 1 :
        masterID = data.loc[data['Account'] == orga, 'AccountID'].mode().min()
        data.loc[data['Account'] == orga, 'MasterID'] = masterID
        domain = data.loc[data['Account'] == orga, 'Email domain'].mode().min()
        data.loc[(data['Account'] == orga) & (data['Email domain'].isna()), 'Email domain'] = domain
        acronym = data.loc[data['Account'] == orga, 'Acronym'].mode().min()
        data.loc[(data['Account'] == orga) & (data['Acronym'].isna()), 'Acronym'] = acronym
```

```python
domains = list(data['Email domain'].unique())
domains
```

```python
for domain in domains :
    if data.loc[data['Email domain'] == domain].shape[0] > 1 :
        masterID = data.loc[data['Email domain'] == domain, 'AccountID'].mode().min()
        data.loc[data['Email domain'] == domain, 'MasterID'] = masterID
        acronym = data.loc[data['Email domain'] == domain, 'Acronym'].mode().min()
        data.loc[(data['Email domain'] == domain) & (data['Acronym'].isna()), 'Acronym'] = acronym
```

```python
acronyms = list(data['Acronym'].unique())
acronyms
```

```python
for acronym in acronyms :
    if data.loc[data['Acronym'] == acronym].shape[0] > 1 :
        masterID = data.loc[data['Acronym'] == acronym, 'AccountID'].mode().min()
        data.loc[data['Acronym'] == acronym, 'MasterID'] = masterID
```
